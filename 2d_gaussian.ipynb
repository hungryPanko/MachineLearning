```python
import numpy as np
import matplotlib.pyplot as plt

# This is similar to the code from lecture 1
# to sample from a 2D Gaussian Distribution
mean = [0,0]
cov = [[1, 1.5], [1.5, 5]]
# Fix random seed to get consistent results
np.random.seed(1024)
X = np.random.multivariate_normal(mean, cov, 1000)
fig, ax = plt.subplots(figsize=(10, 10))
# c='r', dot color is red
# s=10.0, dot size is 10
# alpha=0.3, dot opacity is 0.3
ax.scatter(X[:,0], X[:,1], c='r', s=10.0, alpha=0.3, label="2D-Gaussian")
ax.grid()
ax.legend(loc = 0)
# Set x/y axis limits
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
fig.show()
```

```python
# 3(a) TODO: Report the estimate mean and covariance
# of the sampled points
print("mean_x_axis: ", np.mean(X[:,0]))
print("mean_y_axis: ", np.mean(X[:,1]))
print("Covariance: \n", np.cov(X[:,0],X[:,1])[0][1])
```

```python
# 3(b) TODO: Plot the histogram for the x-coordinates of X
# and y-coordinates of X respectively.
# You can use the plt.hist() function

#x-coordinates of X
plt.title("x-coordinates")
plt.hist(X[:,0], bins=30, color='orange')
plt.show()

#y-coordinates of X
plt.title("y-coordinates")
plt.hist(X[:,1], bins=30, color='red')
plt.show()
```

```python
# 3(c) TODO: Are the x-coordinates of X samples from
# some Gaussian distribution?
# If so, estimate the mean and variance.
# Do the same for the y-coordinates.
print("Yes, both x-coordinates and y-coordinates do follow Gaussian distrubution.")
mean_x = np.mean(X[:,0])
var_x = np.var(X[:,0])
print(" mean of x-cooridnates: ", mean_x)
print(" variance of x-cooridnates: ", var_x)
mean_y = np.mean(X[:,1])
var_y = np.var(X[:,1])
print(" mean of y-cooridnates: ", mean_y)
print(" variance of y-cooridnates: ", var_y)

# 3(d) TODO: Generate a new 2D scatter plot of 1000 points,
# such that the x-coordinates(y-coordinates) of all the points
# are samples from a 1D Gaussian distribution 
# using the estimated mean and variance based on the x-coordinates(y-coordinates) of X.

X_ = np.random.normal(mean_x, np.sqrt(var_x), 1000)
Y_ = np.random.normal(mean_y, np.sqrt(var_y), 1000)
fig, ax = plt.subplots(figsize=(10, 10))
ax.scatter(X_, Y_, c='b', s=10.0, alpha=0.3, label="2D-Gaussian")
ax.grid()
ax.legend(loc = 0)
# Set x/y axis limits
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
fig.show()

print("\nQ: Comparing to original poinrts, wha is the difference about their distribution? What cuases the difference?")
print("A: Mean and Variance stay almost the same. However, covariance information is lost. Thus, Sampled points are not diagonal (skewed) shape anymore. ")
```

```python
# Back to the original X
fig, ax = plt.subplots(figsize=(10, 10))
# c='r', dot color is red
# s=10.0, dot size is 10
# alpha=0.3, dot opacity is 0.3
ax.scatter(X[:,0], X[:,1], c='r', s=10.0, alpha=0.3, label="2D-Gaussian")
ax.grid()
ax.legend(loc = 0)
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])

# 3(e) TODO: Plot a line segment with x = [-10, 10]
# and y = 3x + 1 onto the 2D-Gaussian plot.
# The np.linspace() function may be helpful.
line_seg_x = np.linspace(-10, 10, num=100)
line_seg_y = 3*line_seg_x + 1
ax.plot(line_seg_x, line_seg_y, color='b', label="line segment")
fig.show()
```

```python
# 3(e) TODO: Project X onto line y=3x + 1
# and plot the projected points on the 2D space.
# You need to remove this line and assign the projected points to X_proj.

# dot in X: (a,b)
# dot on the line: (x', 3x'+1)
# slope of line btw (a,b) and dot in the line (3x'+1-b)/(x'-a), slop of y=3x+1 is 3.
# Perpendicular line : m1*m2 = -1
# (3x'+1-b)/(x'-a) * 3 = -1 --> x'=(a+3b-3)/10, y'=3x'+1
X_proj = np.zeros(X.shape)
X_proj[:,0] = (X[:,0]+3*X[:,1]-3)/10
X_proj[:,1] = 3*X_proj[:,0]+1

# Here we only plot 30 points to check the correctness
fig, ax = plt.subplots(figsize=(10, 10))
ax.scatter(X_proj[:30,0], X_proj[:30,1], c='b', s=10.0, alpha=1.0, label="Projected 2D-Gaussian")
ax.scatter(X[:30,0], X[:30,1], c='r', s=10.0, alpha=1.0, label="2D-Gaussian")

ax.grid()
ax.legend(loc = 0)
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
# You can also add the line from the previous plot for verification.
ax.plot(line_seg_x, line_seg_y, color='y', label="line segment")
fig.show()
```

```python
# 3(f) TODO: Draw the histogram of the x-coordinates
# of the projected points.
# Are the x-coordinates of the projected points
# samples from some Gaussian distribution?
# If so, estimate the mean and variance.

#x-coordinates of X_proj
plt.title("x-coordinates")
plt.hist(X_proj[:,0], bins=30, color='orange')
plt.show()

print("Yes, x-coordinates do follow Gaussian distrubution.")
mean_x = np.mean(X_proj[:,0])
var_x = np.var(X_proj[:,0])
print(" mean of x-cooridnates: ", mean_x)
print(" variance of x-cooridnates: ", var_x)
```
